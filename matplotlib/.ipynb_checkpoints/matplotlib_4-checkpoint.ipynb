# MATPLOTLİB NESNE YÖNELİMLİ ÇALIŞMA

```python
#önce figure nesnesi yaratıyoruz  ardından da  bir axis nesnesi  oluşturuyoruz ve  global methodları axis nesnesi üzerinde kullanıyoruz
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.sin(x)

fig = plt.figure(facecolor='yellow',figsize=(15,5)) #figsize ile nesnemizin genişliğini ayarlıyoruz
axis = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axis.plot(x, y)
```

```python
#subpolots kullanımı 

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.sin(x)

fig, axis = plt.subplots(3, 2) #3 satır iki sutunluk subplot alanı yaratıyorum 

axis[0, 0].plot(x, y) #0,0 alanına çizimimi yapıyorum

```

```python
#figure ve axis nesnelerini bir arada yaratıp daha sonradan bu elemanları set edebiliriz 
#figure  nesnemizin attributeleri görünümü değiştiren propertyler  içeririr  
#axis nesnemiz iede eksenlere ait detayları set edebiliriz

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.sin(x)

fig, axis = plt.subplots()

fig.set_facecolor('gray')
fig.set_figwidth(10)

axis.set_title('Sinüs Grafiği')
axis.set_xlabel('x')
axis.set_ylabel('y')
axis.plot(x, y)

axis.plot(x, y)
```

```python
#OOP yöntemiyle histogram örnek kulanımı: 
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)

fig, axis = plt.subplots()

fig.set_figwidth(10)


axis.set_title('Sinüs Grafiği')
axis.set_xlabel('x')
axis.set_ylabel('frequency')
axis.hist(x)
```

```python
#yukarıdaki örneğin gobal yazım biçimi
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)

fig = plt.Figure(figsize=(5, 5))
plt.title('Histogram')
plt.xlabel('x')
plt.ylabel('frequency')
plt.hist(x)
```
