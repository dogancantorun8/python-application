```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = np.sin(x)
plt.title('Sinüs Grafiği')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axis([-20, 20, -3, 3])
plt.plot(x, y, 'o-g')
```

# Grafik Çizim Methodları ve Parametre kullanımı

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('Sinüsve Kosinüs Grafikleri')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axis([-10, 10, -1, 1])
plt.xticks(range(-10, 10 + 1))# x eksenini 10 eşit parçaya ayırdım her adımın arasında 1 değer fark var
plt.yticks(np.arange(-1, 1.1, 0.1)) #y eksenini -1 ve 1 arasında 0.1 adım aralıklarla böldüm
plt.plot(x, y1, 'o-r', label='Sinüs',linestyle='--', color='magenta',marker='o', markersize=5, markerfacecolor='blue') #grafik üzerindeki noktalarımı işaratleyip renklendirdim
plt.plot(x, y2, 'o-b', label='Cosinüs', linewidth=3) #linewidth methoduyla cos eğrimi kalınlaştırdım
plt.legend(['Sinüs', 'Kosinüs'], loc='upper left',edgecolor='red', title='legends', facecolor='gray') #legend fonksiyonun özelliklerini ekledim
```

```python
#Aynı grafik üzerinde iki fonksiyonu çizdirmek istersem
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 30)
y1 = x * x
y2 = x ** 3 - 2 * x ** 2 + 1

plt.plot(x, y1, '-k', x, y2, 'o:r') #x ve y1 den oluşan basit bir eğriyken x ve y2 den oluşan hiperbolik bir eğri oluyor
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.random.randint(0, 100, 5)
y = np.random.randint(0, 100, 5)

plt.scatter(x, y, s=100, c=['red', 'green', 'blue', 'yellow', 'magenta'], marker='o', edgecolor='blue') #her noktanın farklı bir renk ile marker etmek istersem dizi ile tanımlıyorum
```

# Bar oluşturmak:

```python
import numpy as np
import matplotlib.pyplot as plt

x = ['Elma', 'Armut', 'Çilek', 'Muz', 'Portakal']
y = [14, 10, 5, 7, 23]

plt.title('Meyve Satışları')
plt.xlabel('Meyveler', fontsize=14)
plt.ylabel('Satış Miktarları', fontsize=14)
plt.yticks(range(0, 25, 2))
plt.bar(x, y, color='magenta', edgecolor='blue', width=0.8, align='center', linewidth=3)
```
