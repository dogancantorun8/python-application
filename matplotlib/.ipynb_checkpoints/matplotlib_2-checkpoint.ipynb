# Csvden okuyup  matplotlib ile görselleştirme:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'Covid-19-Datasets\covid_19_data.csv') #read csv methoduyla okuma işlemi yaptım
df = df[['Confirmed', 'Deaths', 'Recovered']] # 3 tane kolonu aldım
means = df.mean() # 3kolonun ortalamasını aldım 
plt.title('Covid-19 Death Graph', fontsize=16) 
#barchart şeklinde görselleştiriyorum
plt.bar(means.index, means.values, color=['blue', 'red', 'yellow'], width=0.5) #dizi halinde istediğim renkleri veriyorum 
```

# Histogram Grafiği 


```python
#histogram:Sıklık grafiklerine histogram denir.Her iki eksende sayısaldır.Aralık belirtilir ve o aralıklardaki değer miktarı görselleştirilir 
#histogram çizdiğimde bir durumun olasılık dağılımı hakkında yorum yapabilirim 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'Covid-19-Datasets\covid_19_data.csv')
x = np.random.randn(10000)
plt.title('Normal Random Numbers Histogram', fontsize=15)
plt.xlabel('x', fontsize=12)
plt.ylabel('frequency', fontsize=12)
plt.xticks(np.arange(-10, 10, 0.5))
plt.hist(x, bins=20) 
```

```python
#birden fazla veri setinin histogramını çizmek istersem :
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'Covid-19-Datasets\covid_19_data.csv')
x = np.random.randn(2000)
plt.title('Normal Random Numbers Histogram', fontsize=15)
plt.xlabel('x', fontsize=12)
plt.ylabel('frequency', fontsize=12)
plt.xticks(np.arange(-10, 10, 0.5))
plt.hist(x, rwidth=1, edgecolor='red', color='yellow') #edge colour ile her barın arasındaki çizgileri belli ettim
```

```python
#pasta grafiğinde explode methodu ile istediğimiz dilimi pastadan koparabiliriz 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#df = pd.read_csv(r'Covid-19-Datasets\covid_19_data.csv')
x = [360, 500, 180, 900, 1300]
labels = ['XRP', 'BTC', 'Avax', 'NEO', 'BTG']
plt.pie(x, labels=labels, colors=['red', 'green', 'blue', 'yellow', 'magenta'], explode=[0.1, 0, 0, 0, 0])
```

```python
#pasta dilimlerin üzerinde yüzdeler olsun istersem autopct kullanabilirim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = [664812242762, 141428512203, 24330211680, 15364089757, 12576793068]
labels = ['BTC', 'ETH', 'USDT', 'DOT', 'XRP']
plt.pie(x, labels=labels, explode=[0, 0, 0, 0, 0], autopct='%%%.0f') #yüzdelerinin grafiğin üzerine yazdırdıms
```

```python
#pastanın yarıçapını değiştirip grafiği büyütebilirim :matplotta birimler pixel değil inç cinsindedendir
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = [664812242762, 141428512203, 24330211680, 15364089757, 12576793068]
labels = ['BTC', 'ETH', 'USDT', 'DOT', 'XRP']
plt.pie(x, labels=labels, explode=[0, 0, 0, 0, 0], autopct='%%%.0f',radius=2)
```

```python
#bir figure boyutunu inç cinsinden değiştirebiliriz 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = [664812242762, 141428512203, 24330211680, 15364089757, 12576793068]
labels = ['BTC', 'ETH', 'USDT', 'DOT', 'XRP']
plt.figure(figsize=(10, 10)) #figure parametresiyle config yaptım
plt.pie(x, labels=labels, explode=[0, 0, 0, 0, 0], autopct='%%%.0f')
```

# SUBPLOT ÇİZİM

```python
#Bir figure üzerinde birden fazla çizim yapabilirim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100) #100 tane nokta aldım
y1 = x ** 2
y2 = np.sin(x)
y3 = np.cos(x)
y4 = np.exp(x)

plt.figure(figsize=(10, 10))

plt.subplot(221) #22 ifadesi 2 satır 2 sütunluk 4 eksen yarattım demek ve "1"  4 eksenden ilk yarattığıma çiziyorum anlamına geliyor
plt.title('Sinüs')
plt.plot(x, y1)

plt.subplot(222) #22 ifadesi 2 satır 2 sütunluk 4 eksen yarattım demek ve "2"  4 eksenden ikinci yarattığıma çiziyorum anlamına geliyor
plt.title('X Kare')
plt.plot(x, y2)

plt.subplot(223) #22 ifadesi 2 satır 2 sütunluk 4 eksen yarattım demek ve "2"  4 eksenden üçüncü yarattığıma çiziyorum anlamına geliyor
plt.title('Kosinüs')
plt.plot(x, y3)

plt.subplot(224) #22 ifadesi 2 satır 2 sütunluk 4 eksen yarattım demek ve "2"  4 eksenden dördüncü yarattığıma çiziyorum anlamına geliyor
plt.title('Exp')
plt.plot(x, y4)

plt.savefig('test.png') #yarattığım 4 grafiği locale save ediyorum
```

# Kod bloğunda grafiklerin farklı figurelerde Çizilmesi

```python
#Bunu iki şekilde yapabiliriz ya pause veririz yada farklı bir figure penceresi açıp ikisinin aynı pencerede olmasını engelleyebiliriz
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y1 = x ** 2
y2 = np.sin(x)
y3 = np.cos(x)
y4 = np.exp(x)

plt.plot(x,y1)
plt.pause(1)  ##yeni figure icin pause verilebilir ya da yeni fig acilabilir
plt.plot(x,y2)
plt.figure() #yeni figure açtım
plt.plot(x,y3)
```
