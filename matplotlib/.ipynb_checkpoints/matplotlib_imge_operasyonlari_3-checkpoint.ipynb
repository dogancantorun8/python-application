# Rotate operasyon yapmak istersem 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

img = plt.imread('lena.png')

plt.imshow(img)
height = img.shape[0]
newimg = np.zeros_like(img)  #numpy olduğu için aynı shapede yeni bir numpy nesnesi yaratıyorum

for i in range(height):
    newimg[i] = img[height - 1 - i]
plt.imshow(newimg)
```

```python
#RED olan sütunları sıfırlamak istersem
```

```python
newimg[:,:,0] = 0
plt.imshow(newimg)
```

```python
#2.yol olarak copy bir nesne yaratıü aynı işlemi yapabilirdik 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

img = plt.imread('lena.png')

plt.imshow(img)
height = img.shape[0]
newimg = np.zeros_like(img)

for i in range(height):
    newimg[i] = img[height - 1 - i]
plt.imshow(newimg)
    

img2 = np.copy(img)
img2[:, :, 0] = 0
plt.imshow(img2)
```

# transpoze uygularsam :

```python
img = plt.imread('lena.png')

newimg = np.transpose(img, axes=[1,0,2])
plt.imshow(newimg)
plt.imsave('lena_kayit.jpg', newimg)
```
