# grafiklere legend eklemek istersem 

```python
#legend plotlardan sonra yazılmalı 
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)


plt.title('Sinüsve Kosinüs Grafikleri')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axis([-10, 10, -1, 1])
plt.plot(x, y1, 'o-r',label='Sinüs')
plt.plot(x, y2, 'o-b')
plt.legend(['Sinüs', 'Kosinüs'], loc='upper left')
```

```python
#eksenlerdeki tickleri yani artışları ayarlamak istersem 
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('Sinüsve Kosinüs Grafileri')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axis([-10, 10, -1, 1])
plt.xticks(range(-10, 10 + 1))
plt.plot(x, y1, 'o-r', label='Sinüs')
plt.plot(x, y2, 'o-b', label='Cosinüs')
plt.legend(loc='upper left', edgecolor='red', title='legends', facecolor='gray')
```

# Chart grafiği çizmek istersem

```python
import numpy as np
import matplotlib.pyplot as plt

x = ['Elma', 'Armut', 'Çilek', 'Muz', 'Portakal']
y = [14, 10, 5, 7, 23]

plt.title('Meyva Satışları')
plt.xlabel('Meyvalar', fontsize=14)
plt.ylabel('Satış Miktarları', fontsize=14)
plt.yticks(range(0, 25, 2))
plt.bar(x, y, color='yellow', edgecolor='blue', width=0.8, align='center', linewidth=3)

```

```python
#iki grafikli örneğim:tüm çizim özelliklerini kullandım
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('Sinüsve Kosinüs Grafileri')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axis([-10, 10, -1, 1])
plt.xticks(range(-10, 10 + 1), color='red')
plt.yticks(np.arange(-1, 1.1, 0.1), fontsize=8)
plt.plot(x, y1, label='Sinüs', linestyle='--', color='magenta',marker='o',markerfacecolor='cyan')
plt.plot(x, y2, label='Cosinüs', linewidth=3)
plt.legend(loc='upper left', edgecolor='red', title='legends', facecolor='gray')
```

# Görselleştirilmek istenen verilerden birinin vektör formunda olması

```python
#matrissel biçimde olan verilerde kullanılıabilir
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = np.sin(x).reshape(-1, 1)#reshape yaparak 2 boyutlu hale getirdim
y = np.insert(y, 1, np.cos(x), axis=1)

plt.plot(x, y)
```
